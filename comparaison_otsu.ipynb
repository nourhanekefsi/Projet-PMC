```python
import matplotlib.pyplot as plt

# Facteurs de redimensionnement
resize_factors = [1, 5, 10, 15, 20, 25, 30]

# Temps d'exécution séquentiels (en secondes)
sequential_times = [0.016285, 0.301039, 1.086638, 2.580812, 4.790000, 7.068308, 10.409259]

# Temps d'exécution parallèles (en millisecondes)
parallel_times = [0.843, 14.046, 55.914, 120.444, 168.554, 265.408, 375.300]

# Conversion des temps parallèles en secondes pour les rendre comparables
parallel_times_seconds = [t / 1000 for t in parallel_times]

# Création du graphique
plt.figure(figsize=(10, 6))

# Affichage des temps séquentiels
plt.plot(resize_factors, sequential_times, label='Temps Séquentiel (en sec)', marker='o', color='blue')

# Affichage des temps parallèles
plt.plot(resize_factors, parallel_times_seconds, label='Temps Parallèle (en sec)', marker='x', color='red')

# Ajout des labels, du titre et de la légende
plt.xlabel('Facteur de Redimensionnement')
plt.ylabel('Temps d\'Exécution (en secondes)')
plt.title('Comparaison des Temps d\'Exécution - Otsu Séquentiel vs Parallèle')
plt.legend()

# Affichage de la grille
plt.grid(True)

# Affichage du graphique
plt.show()
```
